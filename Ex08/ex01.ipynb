Q: How many weights and biases does the neuronal network have?
<br> A: Assume hidden layer consists of m nodes, the ouput layer of k nodes and we have n inputs. Then we have n*m+m*k weights and m+k biases.

### b,c,d,e,f)
Implementing neuronal network
Starting with initialization with given test case

```python
import numpy as np
import neuronal_network

input_num = 2
hidden_num = 1
output_num = 1
value = 2
input = np.array([3,2])
exact_output = -2
```

Initialization network

```python
network_value = neuronal_network.init_network_value(input_num, hidden_num, output_num,value)
print(network_value)
```

Forward pass

```python
z,x= neuronal_network.forward_pass(network_value,input)
print("pre activation values: ",z)
print("post activation values: ",x)
```

Backward pass

```python
delta = neuronal_network.backward_pass(network_value,z,x,-2)
calc = neuronal_network.calc_derivatives_loss_function(input,network_value,delta,x)
print(calc)
```

Second network but random initialization, higher dimensions

```python
input_num = 10
hidden_num = 6
output_num = 2
value = 2
input = np.random.rand(1,input_num)
print("input value: ",input)
exact_output = np.random.rand(1,2)
print("output exact: ", exact_output)
```

weightings and biases

```python
network_random = neuronal_network.init_network_random(input_num,hidden_num,output_num)
print(network_random)
```

forward pass

```python
z,x = neuronal_network.forward_pass(network_random,input)
print("pre activation values: ", z)
print("post activation values: ", x)
```

backward pass

```python
delta = neuronal_network.backward_pass(network_random,z,x,exact_output)
calc = neuronal_network.calc_derivatives_loss_function(input,network_random,delta,x)
print(calc)
```
